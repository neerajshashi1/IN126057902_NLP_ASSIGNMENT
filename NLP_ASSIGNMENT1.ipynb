# TASK 2: PREPROCESS FUNCTION

```python
def preprocess_text(text):
    try:
        # Handle empty input
        if not text or text.strip() == "":
            return [], ""

        # Convert to lowercase
        text = text.lower()

        # Remove URLs
        text = re.sub(r"http\S+|www\S+", "", text)

        # Remove emails
        text = re.sub(r"\S+@\S+", "", text)

        # Remove numbers
        text = re.sub(r"\d+", "", text)

        # Remove emojis (basic)
        text = re.sub(r"[^\w\s]", " ", text)

        # Handle repeated characters (soooo → soo)
        text = re.sub(r'(.)\1{2,}', r'\1\1', text)

        # Remove extra spaces
        text = re.sub(r"\s+", " ", text).strip()

        # Tokenization
        tokens = text.split()

        # Remove short words (<=2) except 'no', 'not'
        cleaned_tokens = []
        for word in tokens:
            if len(word) > 2 or word in ['no', 'not']:
                cleaned_tokens.append(word)

        # Final cleaned sentence
        cleaned_sentence = " ".join(cleaned_tokens)

        return cleaned_tokens, cleaned_sentence

    except Exception as e:
        print("Error:", e)
        return [], ""
```

# TASK 3: STRESS TESTING

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print("\nTASK 3: STRESS TESTING\n")

all_tokens = []

results = []

for sentence in test_sentences:
    tokens, clean_sentence = preprocess_text(sentence)

    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean_sentence)
    print("-" * 50)

    all_tokens.extend(tokens)

    results.append({
        "original": sentence,
        "tokens": tokens,
        "clean": clean_sentence
    })
```

#  TASK 4: TOKEN ANALYTICS

```python
print("\nTASK 4: TOKEN ANALYTICS\n")

max_noise = ""
max_tokens = 0

for r in results:
    tokens = r["tokens"]

    if len(tokens) == 0:
        avg_len = 0
    else:
        avg_len = sum(len(w) for w in tokens) / len(tokens)

    print("Sentence:", r["original"])
    print("Total Tokens:", len(tokens))
    print("Unique Tokens:", len(set(tokens)))
    print("Avg Token Length:", round(avg_len, 2))
    print("-" * 40)

    # Most meaningful sentence
    if len(tokens) > max_tokens:
        max_tokens = len(tokens)
        max_noise = r["original"]

print("\nSentence with most meaningful tokens:", max_noise)
```

# TASK 6: FULL PIPELINE

```python


def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

# Run pipeline
pipeline_output = full_pipeline(test_sentences)

print("\nTASK 6: FULL PIPELINE OUTPUT\n")
print(pipeline_output)


     
```
